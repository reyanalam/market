```python
import pandas as pd
```

```python
df = pd.read_csv(r'C:\Users\Reyan\Desktop\New folder\Table_5.13.csv')
```

```python
df.head(10)
```

```python
fuel = df.iloc[3:,1:]
```

```python
fuel.head()
```

```python
fuel['mean'] = fuel.iloc[:,1:7].mean(axis=1)
```

```python
fuel.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.plot(fuel.index, fuel['mean'], marker='o', linestyle='-', color='b', label='Row Mean')
plt.xlabel('Row Index')
plt.ylabel('Mean Value')
plt.title('Mean of Each Row')
plt.legend()
plt.grid(True)
plt.show()
```

```python
fuel_sorted = fuel.sort_values(by='mean', ascending=False)

# Select top 10 and last 10 states
df_top10 = fuel_sorted.iloc[2:13,:]
df_last10 = fuel_sorted.tail(10)

# Plot Top 10
plt.figure(figsize=(8, 4))
plt.bar(df_top10['Sector/ State/UT'], df_top10['mean'], color='green', alpha=0.7)
plt.title('Top 10 States by average fuel consumption')
plt.xlabel('State')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()

# Plot Last 10
plt.figure(figsize=(8, 4))
plt.bar(df_last10['Sector/ State/UT'], df_last10['mean'], color='red', alpha=0.7)
plt.title('Bottom 10 States by average fuel consumption')
plt.xlabel('State')
plt.ylabel('Mean Value')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
